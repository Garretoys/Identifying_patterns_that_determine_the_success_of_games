### Шаг 2. Подготовка данных <a id="step2"></a>  

#### Приведем названия столбцов к нижнему регистру <a id="step2.1"></a>  

```python
columns = df.columns
columns = columns.str.lower()
df.set_axis(columns, axis = 'columns', inplace = True)
df.head()
```

#### Подсчитаем суммарные продажи и выведем в отдельный столбец <a id="step2.2"></a>

```python
df['total_sales'] = df['na_sales'] + df['eu_sales'] + df['jp_sales'] + df['other_sales']
```

Рейтинг tbd расшифровывается как To Be Determined ("Будет определено").<br/>
Значит рейтинг данного продукта по какой-то причине пока не определен.

```python
df[df['user_score'] == 'tbd'].head(20)
```

Вероятно, рейтинг не определен, по причине малого количества отзывов. <br/>
Количество отзывов должно напрямую зависеть от количества проданных копий.<br/>
Попробуем сравнить средние показатели по столбцу total_sales. 

```python
df[df['user_score'] == 'tbd']['total_sales'].mean()
```

```python
df[df['user_score'] != 'tbd']['total_sales'].mean()
```

Среднее количество проданных копий игр, для которых не получилось определить рейтинг, более чем в три раза ниже.<br/>
Из всего этого следует, что tbd можно заменить на NaN.

```python
df.loc[df['user_score'] == 'tbd', 'user_score'] = 'NaN'
```

#### Преобразуем типы данных <a id="step2.3"></a>

```python
df.info()
```

Посмотрим уникальные значения в столбцах, которые возможно потребуют преобразования

```python
df['critic_score'].unique()
```

```python
df['user_score'].unique()
```

Преобразуем данные в следующих столбцах: 
- year_of_release - в int, год ведь не может быть дробным?
- critic_score - значения только целочисленные, но для выведения корреляции с общим количество продаж нам нужно, чтобы обе колонки были в едином формате, поэтому менять тип не будем.
- user_score - а тут наоборот, значения не натуральные, столбец должен быть во float

```python
df['year_of_release'] = df['year_of_release'].astype('Int64')
df['user_score'] = df['user_score'].astype('float')
```

#### Влияет ли рейтинг ESRB на продажи в отдельном регионе? <a id="step4.3"></a> 

Информация по рейтингам:
1. «EC» («Early childhood») — «Для детей младшего возраста»: Игра подходит для детей от 3 лет и старше и не содержит материалов, которые родители могли бы счесть неподходящими. Продукты, получившие данный рейтинг, изначально разрабатываются для детей и обычно представляют собой развивающие игры. Некоторые усложнённые развивающие игры могут иметь рейтинг «Everyone».
2. «E» («Everyone») — «Для всех»: Содержание вполне подходит для возрастной категории от 6 лет и старше; такие игры могут понравиться и взрослым. Игры с этим рейтингом могут содержать минимальное насилие, в основном «мультяшного» характера. Первой игрой, которая получила данный рейтинг, стала The Simpsons Cartoon Studio, выпущенная в 1996 году. Первоначально "K-A" ("Kids to Adults").
3. «E10+» («Everyone 10 and older») — «Для всех от 10 лет и старше»: Проекты с данным рейтингом могут содержать немного больше мультипликационного или мягкого насилия, или несколько откровенные сцены или минимальное количество крови. Рейтинг был принят ESRB 2 марта 2005 года. Первой игрой, которая получила данный рейтинг, стала Donkey Kong: Jungle Beat.
4. «T» («Teen») — «Подросткам»: Игра подходит для лиц от 13 лет и старше. Проекты из данной категории могут содержать насилие, непристойные сцены, грубый юмор, в меру откровенное сексуальное содержимое, кровь или нечастое использование ненормативной лексики.
5. «M» («Mature») — «Для взрослых»: Материалы игры не подходят для подростков младше 17 лет. Проекты с данным рейтингом могут содержать достаточно жестокое насилие, большое количество крови с расчленением, непристойные сексуальные сцены или грубую ненормативную лексику, нежелательную для младшей аудитории. 
6. «AO» («Adults Only 18+») — «Только для взрослых»: Содержание игры только для взрослых старше 18 лет. Продукты из данной категории могут содержать длительные сцены жестокого насилия и/или очень откровенное сексуальное содержимое, а также сцены с обнажением. Большинство таких игр предназначены для персональных компьютеров под управлением Microsoft Windows и Apple Macintosh. Рейтинг «Только для взрослых» является предметом многочисленных дискуссий, так как накладывает серьёзные ограничения на продажи игры.
7. «RP» («Rating Pending») — «Рейтинг ожидается»: Продукт был отправлен в ESRB и ожидает присвоения рейтинга. Данный логотип используется только на рекламных презентациях и в демо-версиях игр до официальной даты выпуска в продажу.

### Шаг 5. Проверка гипотез<a id="step5"></a> 

```python
from scipy import stats as st
```

#### Гипотеза №1<a id="step5.1"></a>
Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Сформулируем нулевую и альтернативную гипотезы:
- Нулевая. Средние пользовательские рейтинги платформ Xbox One и PC **равны**.
- Альтернативная. Средние пользовательские рейтинги платформ Xbox One и PC **отличаются**.

```python
adf_us_dropna = adf.dropna(subset=['user_score'])
pc = adf_us_dropna[adf_us_dropna['platform'] == 'PC']['user_score']
xone = adf_us_dropna[adf_us_dropna['platform'] == 'XOne']['user_score']
print('Дисперсия для выборки по PC равна:', np.var(pc, ddof=1))
print('Дисперсия для выборки по Xbox One равна:', np.var(xone, ddof=1))
```

```python
alpha = .05 # критический уровень статистической значимости
            # если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(
    pc, 
    xone, equal_var=False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Заключение. Подтвердили первую гипотезу, средние пользовательские рейтинги платформ Xbox One и PC **могут быть равны**.

#### Гипотеза №2<a id="step5.2"></a>
Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Сформулируем нулевую и альтернативную гипотезы:
- Нулевая. Средние пользовательские рейтинги жанров action и sports **равны**.
- Альтернативная. Средние пользовательские рейтинги жанров action и sports **отличаются**.

```python
action = adf_us_dropna[adf_us_dropna['genre'] == 'Action']['user_score']
sports = adf_us_dropna[adf_us_dropna['genre'] == 'Sports']['user_score']
print('Дисперсия для выборки по Action равна:', np.var(action, ddof=1))
print('Дисперсия для выборки по Sports равна:', np.var(sports, ddof=1))
```

```python
alpha = .05 # критический уровень статистической значимости
            # если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(
    action, 
    sports, equal_var=False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Заключение. Подтвердили вторую гипотезу, средние пользовательские рейтинги жанров Action и Sports вероятнее всего **разные**.

5. Action <br/>
Жанр экшн занял первое место и по количеству игр созданных за последние 10 лет, и в топ-5 по количеству проданных копий, но не попал даже в топ-5(шестое место) по среднему показателю продаж. Не самый интересный жанр, для рекламных акций. <br/>
    
При проведении **разных рекламных кампаний** в разных регионах, и при условии рекламирования конкретного продукта, нужно придерживаться следующих рейтингов:
 - Северная Америка:
     1. Shooter
     2. Platform
     3. Sports
     4. Racing
     5. Fighting <br/>
<br/>    
 - Европа:
     1. Shooter
     2. Platform
     3. Racing
     4. Sports
     5. Action <br/>
<br/>   
 - Япония:
     1. Role-Playing
     2. Platform
     3. Fighting
     4. Misc
     5. Action   

Кроме прочего, стоит учитывать рейтинг ESRB, при планировании рекламной кампании в Европе и Японии.<br/>

    Для Европы заинтересованность покупателей в рейтингах расположилась в порядке убывания так:
M, E, E10+, T, RP, EC. <br/>
При степени корреляции 0.15.

    Для Японии заинтересованность покупателей в рейтингах расположилась в следующем порядке:
E, T, M, E10+, RP, EC. <br/>
При степени корреляции 0.22.
    
    В Северной Америке зависимость продаж от рейтинга практически отсутствует.
